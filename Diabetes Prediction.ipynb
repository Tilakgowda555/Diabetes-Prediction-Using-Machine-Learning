# Step1: Importing libraries,loading data and reading data using pandas

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('pima-data.xlsx')
df
```

```python
df.drop('thickness', axis=1,inplace=True) #droping columns
df.drop('has_diabetes', axis=1,inplace=True)
df.drop('diabetes_orig',axis=1,inplace=True)
```

```python
corr = df.corr() #after checking correlation
fig,ax = plt.subplots(figsize=(6,6))
ax.matshow(corr)
plt.xticks(range(len(corr.columns)),corr.columns,rotation=90)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.show()
```

```python
dict1 = {'diabetes':[True,False]} #checking any text data is available or not
dict2 = {'diabetes':[1,0]}
df.replace(dict1,dict2,inplace=True)
df
```

```python
#checking the data is balanced or not
num_true = len(df.loc[df['diabetes']==1])
num_false = len(df.loc[df['diabetes']==0])
print(f'num_true={num_true}')
print(f'num_false={num_false}')
per_num_true = (num_true/(num_true + num_false))*100
print(per_num_true)
per_num_false = (num_false/(num_false + num_true))*100
print(per_num_false)
```

# Step3: Training and spliting

```python
from sklearn.model_selection import train_test_split
input_columns = ['num_preg','glucose_conc','diastolic_bp','insulin','bmi','diab_pred','age','skin']
output_columns = ['diabetes']
x = df[input_columns].values
y = df[output_columns].values
split_test_size = 0.3
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = split_test_size,random_state = 42)
```

```python
print('training true:{0} ({1:0.2}%)'.format(len(y_train[y_train[:]==1]),(len(y_train[y_train[:]==1])/len(y_train)*100)))
print('training false:{0} ({1:0.2}%)'.format(len(y_train[y_train[:]==0]),(len(y_train[y_train[:]==0])/len(y_train)*100)))
print('test true:{0} ({1:0.2}%)'.format(len(y_test[y_test[:]==1]),(len(y_test[y_test[:]==1])/len(y_test)*100)))
print('test false:{0} ({1:0.2}%)'.format(len(y_test[y_test[:]==0]),(len(y_test[y_test[:]==0])/len(y_test)*100)))
```

# Step4: Post split data preparation

```python
#counting the number of rows if zeros is present in each column
print('# rows in dataframe {0}'.format(len(df)))
print('# rows missing num_preg:{0}'.format(len(df.loc[df['num_preg']==0])))
print('# rows missing glucose_conc:{0}'.format(len(df.loc[df['glucose_conc']==0])))
print('# rows missing diastolic_bp:{0}'.format(len(df.loc[df['diastolic_bp']==0])))
print('# rows missing diabetes:{0}'.format(len(df.loc[df['diabetes']==0])))
print('# rows missing insulin:{0}'.format(len(df.loc[df['insulin']==0])))
print('# rows missing bmi:{0}'.format(len(df.loc[df['bmi']==0])))
print('# rows missing diab_pred:{0}'.format(len(df.loc[df['diab_pred']==0])))
print('# rows missing age:{0}'.format(len(df.loc[df['age']==0])))
print('# rows missing skin:{0}'.format(len(df.loc[df['skin']==0])))
```

```python
#handling zeros in our data
from sklearn.impute import SimpleImputer
fill_zeros = SimpleImputer(missing_values=0,strategy = 'mean')
x_train = fill_zeros.fit_transform(x_train)
x_test = fill_zeros.fit_transform(x_test)
```

```python
x_train[0:3]
```

# Step5: Training the model

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(x_train,y_train.ravel())
```

```python
#checking the accuracy of the train model
nb_predict_train = nb_model.predict(x_train)
from sklearn import metrics
print('accuracy : {0:.4f}'.format(metrics.accuracy_score(y_train,nb_predict_train)))
```

# Step6: Testing the model

```python
#checking the accuracy of the train model
nb_predict_test = nb_model.predict(x_test)
from sklearn import metrics
print('accuracy : {0:.4f}'.format(metrics.accuracy_score(y_test,nb_predict_test)))
```

```python
#checking  accuracy by confusion matrix
print('confusion matrix')
print('{0}'.format(metrics.confusion_matrix(y_test,nb_predict_test)))
```

```python
#classification report
print('classification report')
print(metrics.classification_report(y_test,nb_predict_test))
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train,y_train.ravel())
```

```python
#checking the accuracy of the train model
rf_predict_train = rf_model.predict(x_train)
from sklearn import metrics
print('accuracy : {0:.4f}'.format(metrics.accuracy_score(y_train,rf_predict_train)))
```

```python
#checking the accuracy of the test model
rf_predict_test = rf_model.predict(x_test)
from sklearn import metrics
print('accuracy : {0:.4f}'.format(metrics.accuracy_score(y_test,rf_predict_test)))
```

```python
#checking  accuracy by confusion matrix
print('confusion matrix')
print('{0}'.format(metrics.confusion_matrix(y_test,rf_predict_test)))
```

```python
#classification report
print('classification report')
print(metrics.classification_report(y_test,nb_predict_test))
```

```python
#knn algorithm
from sklearn.neighbors import KNeighborsClassifier
k_neighbor = KNeighborsClassifier()
k_neighbor.fit(x_train,y_train.ravel())
```

```python
#checking the accuracy of the train model
x_pred = k_neighbor.predict(x_train)
print(f'Accuracy = {metrics.accuracy_score(y_train,x_pred)}')
```

```python
#checking the accuracy ot the test model
x_pred = k_neighbor.predict(x_test)
print(f'Accuracy = {metrics.accuracy_score(y_test,x_pred)}')
```

```python
#checking accuracy by confusion matrix
print('confusion matrix')
print('{0}'.format(metrics.confusion_matrix(y_test,x_pred)))
```

```python
#checking accuracy by classification report
print('classification report')
print(metrics.classification_report(y_test,x_pred))
```
